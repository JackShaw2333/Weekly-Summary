### 3.2 Bar柱状图

上图是本节我们要实现的目标！分布有上下的柱状图以及文字标识

`np.random.uniform(low=,high=,size=)`从均匀分布$[low,high)$中随机采样

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#n = 12表明“柱”的个数
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.figure()


#“+”表示柱向上
#“-”表示柱向下
plt.bar(X,+Y1)
plt.bar(X,-Y2)




plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#n = 12表明“柱”的个数
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.figure()


#“+”表示柱向上
#“-”表示柱向下
#“#”标识16进制颜色表示
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')



plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#n = 12表明“柱”的个数
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.figure()


#“+”表示柱向上
#“-”表示柱向下
#“#”标识16进制颜色表示
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

#对向上的柱状图加文字
#此处zip(X,Y1)分别将X和Y1的值传入x,y中
for x,y in zip(X,Y1):
    #ha表示水平对齐方式
    #va表示竖直对齐方式
    plt.text(x,y+0.05,'%.2f'%y,ha='center',va='bottom')
    
for x,y in zip(X,Y2):
    #ha表示水平对齐方式
    #va表示竖直对齐方式
    #这里Y2都是正数，所以这里的
    plt.text(x,-y-0.05,'-%.2f'%y,ha='center',va='top')


plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())

plt.show()
```

### 3.3 Contours 等高线图

实现目标如图。

颜色产生从红色到白色的渐变。

```python
#计算高度的公式
def f(x,y):
    return(1-x/2+x**2+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)

#把x,y绑定为网格的输入值
X,Y = np.meshgrid(x,y)

plt.figure()
#cmap:每一个值对应一个颜色的点
#8表示将等高线分为10部分
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)

#画等高线
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)

#加描述
#inline=True表示数值画在线里面;若改为False则线穿过数字
plt.clabel(C,inline=True,fontsize=10)


plt.xticks(())
plt.yticks(())
plt.show()
```

### 3.4 Image图片
本节我们用plt创建图像，但并不是现实物体的图片，如下图所示



```python
#a为我们的伪图片，每个值表示像素
np.random.seed(0)
a = np.random.uniform(0,1,9).reshape(3,3)


plt.figure()

#upper,lower标识颜色按深浅的排列顺序
#interpolation有很多可选值
plt.imshow(a, interpolation='nearest',cmap='bone',origin='upper')

#shrink可以设置高度，0.9表示原高度的90%
#颜色标尺
plt.colorbar(shrink=0.9) 


plt.xticks(())
plt.yticks(())
plt.show()
```

### 3.6 3D数据

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

#在ax上增加3D的axes
ax = Axes3D(fig)

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)

#将X,Y匹配到底面
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

#画图是画在ax上面
#rstride表示行方向的步长，cstride表示列方向的步长
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

#zdir表明投影方向，offset表明投影方向的一个具体平面
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')
ax.set_zlim(-2,2)

plt.show()
```

### 4.1 Subplot 多合一显示

```python
plt.figure()

#subplot(行，列，当前)
#一个plt.subplot()的范围规则与一个plt.figure()的规则相似
plt.subplot(2,2,1)
plt.plot([0,1],[0,1])

plt.subplot(2,2,2)
plt.plot([0,1],[0,2])

plt.subplot(2,2,3)
plt.plot([0,1],[0,3])

plt.subplot(2,2,4)
plt.plot([0,1],[0,4])



plt.show()
```

如果我们想让一张图占更多的位置，我们可以通过修改`plt.subplot()`内的行/列数来实现。

如我们想让1张图占一行，另外3张图占第二行。

```python
plt.figure()

#subplot(行，列，当前)
#一个plt.subplot()的范围规则与一个plt.figure()的规则相似
plt.subplot(2,1,1) #2行1列
plt.plot([0,1],[0,1])

plt.subplot(2,3,4) #分成2行3列后，subplot的编码按总数为6来排
plt.plot([0,1],[0,2])

plt.subplot(2,3,5)
plt.plot([0,1],[0,3])

plt.subplot(2,3,6)
plt.plot([0,1],[0,4])



plt.show()
```

不难发现，上述的分割方式非常`Confusing`，下一节我们会学习到一种简单直白的方法。

### 4.2 Subplot 分格显示
下面我们将学习3种不同的推荐使用的方法，比上一节的那个`confusing`的方法好很多。

方法一 `subplot2grid`

方法二 `gridspec`

    应用此方法时，需要import matplotlib.gridspec as gridspec

方法三 `easy to define structure`


#### 方法一

上图中每个坐标图都是一个`ax`

                1
        2
                        3
    4        5


```python
plt.figure()

#subplot2grid就是单元小格
#(3,3)表示分成3x3个最小单元
#(0,0)表示起始最小单元格
#colspan表示列跨度（以最小单元格为单位）
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)
ax1.plot([1,2],[1,2])

#ax的label设置与plt的不同
ax1.set_title('ax1_title')
ax1.set_xlabel('This is x axis')
ax1.set_ylabel('This is y axis')


ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)
ax3 = plt.subplot2grid((3,3),(1,2),colspan=1,rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,0),colspan=1,rowspan=1)
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1,rowspan=1)

#plt.tight_layout()用于自动调整图像外部的边缘（不能很好的控制图像间的间隔）
plt.tight_layout()

plt.show()
```

#### 方法二
gridspec

需要`import matplotlib.gridspec as gridspec`

```python
import matplotlib.gridspec as gridspec

plt.figure()
gs = gridspec.GridSpec(3,3)
ax6 = plt.subplot(gs[0,:])
ax7 = plt.subplot(gs[1,:2])
ax8 = plt.subplot(gs[1:,2])
ax9 = plt.subplot(gs[2,0])
ax10 = plt.subplot(gs[2,1])

plt.tight_layout()
plt.show()
```

#### 方法三
`plt.subplots()`

```python
#分为2行2列，共享x轴
#((第1行所有的axes),(第2行所有的axes))
f,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,sharex=True,sharey=True)

ax11.scatter([1,2],[1,2])

```
