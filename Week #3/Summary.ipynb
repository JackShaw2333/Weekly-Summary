```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#n = 12表明“柱”的个数
n = 12
X = np.arange(n)
Y1 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.figure()


#“+”表示柱向上
#“-”表示柱向下
#“#”标识16进制颜色表示
plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')

#对向上的柱状图加文字
#此处zip(X,Y1)分别将X和Y1的值传入x,y中
for x,y in zip(X,Y1):
    #ha表示水平对齐方式
    #va表示竖直对齐方式
    plt.text(x,y+0.05,'%.2f'%y,ha='center',va='bottom')
    
for x,y in zip(X,Y2):
    #ha表示水平对齐方式
    #va表示竖直对齐方式
    #这里Y2都是正数，所以这里的
    plt.text(x,-y-0.05,'-%.2f'%y,ha='center',va='top')


plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())

plt.show()
```

### 3.3 Contours 等高线图

实现目标如图。

颜色产生从红色到白色的渐变。

```python
#计算高度的公式
def f(x,y):
    return(1-x/2+x**2+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)

#把x,y绑定为网格的输入值
X,Y = np.meshgrid(x,y)

plt.figure()
#cmap:每一个值对应一个颜色的点
#8表示将等高线分为10部分
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.hot)

#画等高线
C = plt.contour(X,Y,f(X,Y),8,colors='black',linewidth=.5)

#加描述
#inline=True表示数值画在线里面;若改为False则线穿过数字
plt.clabel(C,inline=True,fontsize=10)


plt.xticks(())
plt.yticks(())
plt.show()
```

### 3.6 3D数据

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

#在ax上增加3D的axes
ax = Axes3D(fig)

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)

#将X,Y匹配到底面
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.sin(R)

#画图是画在ax上面
#rstride表示行方向的步长，cstride表示列方向的步长
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

#zdir表明投影方向，offset表明投影方向的一个具体平面
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')
ax.set_zlim(-2,2)

plt.show()
```

### 4.2 Subplot 分格显示
下面我们将学习3种不同的推荐使用的方法，比上一节的那个`confusing`的方法好很多。

方法一 `subplot2grid`

方法二 `gridspec`

    应用此方法时，需要import matplotlib.gridspec as gridspec

方法三 `easy to define structure`


#### 方法一

上图中每个坐标图都是一个`ax`

                1
        2
                        3
    4        5


```python
plt.figure()

#subplot2grid就是单元小格
#(3,3)表示分成3x3个最小单元
#(0,0)表示起始最小单元格
#colspan表示列跨度（以最小单元格为单位）
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)
ax1.plot([1,2],[1,2])

#ax的label设置与plt的不同
ax1.set_title('ax1_title')
ax1.set_xlabel('This is x axis')
ax1.set_ylabel('This is y axis')


ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)
ax3 = plt.subplot2grid((3,3),(1,2),colspan=1,rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,0),colspan=1,rowspan=1)
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1,rowspan=1)

#plt.tight_layout()用于自动调整图像外部的边缘（不能很好的控制图像间的间隔）
plt.tight_layout()

plt.show()
```

#### 方法二
gridspec

需要`import matplotlib.gridspec as gridspec`

```python
import matplotlib.gridspec as gridspec

plt.figure()
gs = gridspec.GridSpec(3,3)
ax6 = plt.subplot(gs[0,:])
ax7 = plt.subplot(gs[1,:2])
ax8 = plt.subplot(gs[1:,2])
ax9 = plt.subplot(gs[2,0])
ax10 = plt.subplot(gs[2,1])

plt.tight_layout()
plt.show()
```

#### 方法三
`plt.subplots()`

```python
#分为2行2列，共享x,y轴
#((第1行所有的axes),(第2行所有的axes))
#f表示figure的值
f,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,)

ax11.scatter([1,2],[1,2])
#因为x,y轴共享，所以右上角的ax没有刻度值

#如果要改figure的参数，通过'f'修改就可以

plt.tight_layout()
plt.show()
```

综合上述3种方法，我个人更喜欢第一种方法，因为在设置参数时简单明了。

### 5.1 Animation 动画

```python
%matplotlib notebook
from matplotlib import animation
from IPython.display import HTML
fig, ax = plt.subplots()

x = np.arange(0,2*np.pi,0.01)
line, = ax.plot(x, np.sin(x))
#输入“，”因为返回值是一个列表，选择第一位，所有加“，”


def animate(i): #i表示第i帧
    line.set_ydata(np.sin(x+i/10))
    return (line,)

def init():
    line.set_ydata(np.sin(x))
    return (line,)

#fig表示图像，func表示动画内的函数，frames表示帧数
#init_func表示初始帧数的样子
#interval表示隔多长时间更新一次，以毫秒为单位。
#blit表示是否更新所有点（False）还是只更新变化了点（True）
ani = animation.FuncAnimation(fig=fig,func=animate,frames=100,init_func=init,interval=20,blit=True)
```

### 5 一个栗子

```python
import tensorflow as tf
import numpy as np

## create data
#tensorflow中大部分数据的类型都是np.float32
x_data = np.random.rand(100).astype(np.float32)
y_data = x_data*0.1 + 0.3

## create tensorflow structure start
#Weights之所以大写是因为可能是矩阵，大写容易辨别
Weights = tf.Variable(tf.random_uniform([1],-1.0,1.0))
biases = tf.Variable(tf.zeros([1]))


y = Weights*x_data + biases

loss = tf.reduce_mean(tf.square(y-y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5) #0.5是学习率
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()
## create tensorflow structure end

sess = tf.Session()
sess.run(init) #这样init就被激活，然后所有变量也被激活

for step in range(201):
    sess.run(train)
    if step%20 == 0:
        print(step,sess.run(Weights),sess.run(biases))
```

备注：
1. `tf.random_uniform(shape,mean,stddev,dtype=tf.float32,seed=None,name=None)`
`shape`用一个list表示tensor的形状

    `mean`均值
    
    `stddev`标准差

### 6 Session会话控制
使用`sess.run()`可以运行创造的结构中的某个图。如果run某个参数的话也会输出参数的值。

先以两个矩阵相乘为例

```python
matrix1 = tf.constant([[3,3]])
matrix2 = tf.constant([[2],
                       [2]])

product = tf.matmul(matrix1,matrix2)

#方法一
sess = tf.Session()
result = sess.run(product)
print(result)
sess.close()

#方法二，这样就无需sess.close()语句，因为出了with后，sess自动被关闭
with tf.Session() as sess:
    result2 = sess.run(product)
    print(result2)
```

### 7 Variable 变量
tensorflow中定义变量与python中有所不同，需要用`tf.Variable()`定义变量

```python
state = tf.Variable(0,name='counter')
print(state.name)
```

定义变量时还可以设置它的name

因为state是第一个变量，所有后面有1个0

```python
one = tf.constant(1)
new_value = tf.add(state,one)
update = tf.assign(state,new_value) #把new_value赋给state

#设置变量后最重要的一步：
init = tf.global_variables_initializer() #初始化所有的变量，这样变量才能被激活

with tf.Session() as sess:
    sess.run(init)
    for _ in range(3):
        sess.run(update)
        print(sess.run(state))
```

### 4 通用学习模式
Sklearn将许多学习模式整合在一起，只要学会一种，就可以做到触类旁通。

下面以在iris数据集上使用knn模型实现分类为例。

```python
from sklearn import datasets
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = datasets.load_iris()
iris_X = iris.data #iris的数据
iris_y = iris.target #iris的标签

#print(iris_X[:2,:])
#print(iris_y)
#花有三种类别

#将数据分类成训练集和测试集，测试比占30%
X_train, X_test, y_train, y_test = train_test_split(
    iris_X,iris_y,test_size=0.3)

#print(y_train)
#可见，y_train相比iris_y而言被打乱了。

knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
#fit后数据已经训练好了。

print(knn.predict(X_test)) #输出对X_test的预测结果
print(knn.score(X_test,y_test)) #输出当前模型在X_test上的效果
```

以`波士顿房价`为例

sklearn的代码相似度非常高

```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression

loaded_data = datasets.load_boston()
data_X = loaded_data.data
data_y = loaded_data.target

model = LinearRegression() #参数可以改变，但默认值就非常好
model.fit(data_X,data_y)

print(model.predict(data_X[:4,:]))
print(model.score(data_X,data_y))

```
